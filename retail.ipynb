```python
import pandas as pd
import scipy.stats as stats

# Assuming you have already loaded your DataFrame and set 'Order Date' as the index
# If not, convert 'Order Date' to datetime and set it as the index (as shown previously)

# Group the data by a specific column, e.g., 'Category', and apply sum to 'Sales' for 2012 and 2013
grouped_2012 = sliced_df_2012
grouped_2013 = sliced_df_2013

# Extract the sales data for 2012 and 2013
sales_2012 = grouped_2012['mean']
sales_2013 = grouped_2013['mean']

# Perform a two-sample t-test
t_statistic, p_value = stats.ttest_ind(sales_2012, sales_2013)

# Set the significance level (alpha)
alpha = 0.05

# Compare the p-value to the significance level
if p_value < alpha:
    print("Reject the null hypothesis: There is a statistically significant difference between 2012 and 2013 sales.")
else:
    print("Fail to reject the null hypothesis: There is no statistically significant difference between 2012 and 2013 sales.")
```

```python
import pandas as pd
from statsmodels.tsa.stattools import adfuller, kpss

# Assuming you have already loaded your DataFrame and set 'Order Date' as the index
# If not, convert 'Order Date' to datetime and set it as the index (as shown previously)

# Extract sales data for 2012 and 2014
sales_2012 = sliced_df_2012_2013['mean']


# Define a significance level (alpha)
alpha = 0.05

# Perform the ADF test for seasonality on 2012 data
adf_result_2012 = adfuller(sales_2012)
adf_p_value_2012 = adf_result_2012[1]


# Perform the KPSS test for seasonality on 2012 data
kpss_result_2012 = kpss(sales_2012, regression='c')
kpss_p_value_2012 = kpss_result_2012[1]

# Interpret and compare the test results
def interpret_test_result(p_value, test_name, year):
    print(f'{test_name} p-value for {year} data: {p_value:.4f}')
    if p_value < alpha:
        print(f'Reject the null hypothesis (H0): {year} data may have seasonality.')
    else:
        print(f'Fail to reject the null hypothesis (H0): {year} data is likely stationary.')

print("2012-2013 Data:")
interpret_test_result(adf_p_value_2012, "ADF", "2012_2013")
interpret_test_result(kpss_p_value_2012, "KPSS", "2012_2013")
```


The test results for the combined 2012-2013 sales data indicate the following:

Augmented Dickey-Fuller (ADF) Test:

ADF p-value for 2012-2013 data: 0.0000
Interpretation: Reject the null hypothesis (H0): 2012-2013 data may have seasonality. The p-value is very low, suggesting strong evidence of seasonality in the combined data.
Kwiatkowski-Phillips-Schmidt-Shin (KPSS) Test:

KPSS p-value for 2012-2013 data: 0.1000
Interpretation: Fail to reject the null hypothesis (H0): 2012-2013 data is likely stationary. The p-value is relatively high, indicating that the combined data is likely stationary around a deterministic trend.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from statsmodels.tsa.arima.model import ARIMA
import statsmodels.api as sm

# Load your time series data into a DataFrame
# Replace 'your_data.csv' with your actual data file
# Assuming your DataFrame has a DateTimeIndex and a 'Sales' column

# Step 1: Split the data into training and test sets
train_size = int(len(sliced_df_2012_2013 ) * 0.8)  # 80% for training
train, test = sliced_df_2012_2013 .iloc[:train_size], sliced_df_2012_2013 .iloc[train_size:]

# Step 2: Calculate the first-order differences (Theta)
train_diff = train['mean'].diff().dropna()

# Step 3: Calculate the average of the first-order differences (Theta)
theta = train_diff.mean()

# Step 4: Forecast using the Theta Method
forecast_values = test.index.map(lambda x: train['mean'].iloc[-1] + theta * (x - train.index[-1]).days)

# Step 5: Evaluate the model's performance (MSE)
mse = mean_squared_error(test['mean'], forecast_values)
print(f"Mean Squared Error (MSE): {mse:.2f}")

# Step 6: Calculate AIC and BIC for the Theta Method
n = len(train) + len(test)
k = 1  # Number of parameters in the Theta Method model (1 for theta)
aic = n * np.log(mse) + 2 * k
bic = n * np.log(mse) + k * np.log(n)

print(f"Akaike Information Criterion (AIC): {aic:.2f}")
print(f"Bayesian Information Criterion (BIC): {bic:.2f}")

# Step 7: Visualize the forecasts
plt.figure(figsize=(12, 6))
plt.plot(train.index, train['mean'], label='Train Data')
plt.plot(test.index, test['mean'], label='Test Data')
plt.plot(test.index, forecast_values, label='Theta Method Forecast', color='red')
plt.title('Sales Forecast using Theta Method')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel

# Load your dataset (replace 'your_data.csv' with your actual data file)
df =df
# Select the important columns for content-based recommendations
important_columns = ['Product ID', 'Product Name', 'Category', 'Sub-Category', 'Sales']

# Create a new DataFrame with only the important columns
df = df[important_columns].head(90000)

# Drop rows with missing values and duplicates
df = df.dropna().drop_duplicates()

# TF-IDF Vectorization
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf_vectorizer.fit_transform(df['Product Name'])

# Compute the cosine similarity between products
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)

# Function to get product recommendations
def get_recommendations(product_name, cosine_sim=cosine_sim, top_n=5):
    # Check if the input product exists in the DataFrame
    if product_name not in df['Product Name'].values:
        return []

    idx = df[df['Product Name'] == product_name].index[0]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:top_n+1]
    product_indices = [i[0] for i in sim_scores]
    return df['Product Name'].iloc[product_indices]
# Example: Get recommendations for a specific product
input_product = 'Tenex Antistatic Computer Chair Mats'  # Replace with the product you want recommendations for
recommendations = get_recommendations(input_product)
print(f"Recommended Products for '{input_product}':")
print(recommendations)
```
